```python
#importing packages

import pandas as pd
import numpy as np
import seaborn as sns
import os
working_directory = os.getcwd()
print(working_directory)

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjustment to plot config


#read files in

df = pd.read_csv('/Users/peter/Downloads/movies.csv')
```

```python
#initial data look
df.head()
```

```python
#missing data identification
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#dropping missing data and check
df = df.dropna()

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))

# data types for columns
df.dtypes
```

```python
#column data type adaptation
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(np.int64)

df['votes'] = pd.to_numeric(df['votes'], errors='coerce').fillna(0).astype(np.int64)

df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(np.int64)
```

```python
df.head()
```

```python
#create year column from release date
#mitigate issues where there is a mismatch
df['yearcorrect'] = df['released'].str.extract(pat ='([0-9]{4})').astype(int)

df.head()
```

```python
df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
#Ability to see all rows
#pd.set_option('display.max_rows', None)
```

```python
#duplicate check and drop

df.drop_duplicates()
```

```python
#scatter plot with Budget vs Gross


plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
#looking at correlation
#3 types, pearson(original), kendall, spearman
# include numeric_only=True parameter to avoid errors of str to float for other cols

df.corr(numeric_only=True) #pearson
```

```python
df.corr(method='kendall',numeric_only=True) #Kendall
```

```python
df.corr(method='spearman',numeric_only=True) #spearman
```

```python
correlation_matrix = df.corr(numeric_only=True) #pearson

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show()
```
